```python
import pandas as pd

df = pd.read_csv('/content/internship_sales_data.csv')
df.head()
```

```python
total_revenue = df['Revenue'].sum()
total_revenue
```

```python
region_sales = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
region_sales
```

```python
region_sales.plot(kind='bar', title='Region-wise Revenue')
plt.ylabel('Revenue')
plt.show()
```

```python
product_sales = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
product_sales
```

```python
product_sales.plot(kind='bar', title='Product-wise Revenue')
plt.ylabel('Revenue')
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x='Quantity', y='Revenue', data=df)
plt.title('Quantity vs Revenue')
plt.show()
```

**Final Insights**

   Key Findings:
- The dataset contains 5 records and 7 features with no missing values.
- The total revenue generated from all sales is ₹317,000.
- The North region generated the highest revenue.
- Laptop is the top revenue-generating product.
- Quantity sold does not directly determine revenue; unit price plays a major role.

  Business Recommendations:
- Focus more on high-value products like laptops and tablets to maximize revenue.
- Improve marketing strategies in low-performing regions.
- Introduce promotions to increase sales quantity without reducing unit price significantly.:::

**Final Conclusion**

This analysis provided insights into sales performance by evaluating revenue trends, product-wise and region-wise sales, and the relationship between quantity and revenue. The findings highlight that high-priced products significantly contribute to total revenue.
